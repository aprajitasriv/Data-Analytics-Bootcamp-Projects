```python
#create empty lists to extract data
#these are list names

idls = [] #ID of the beer
namels = [] #name of the beer
yr_establish = [] 
status = [] 
is_biz = [] 
is_mass = [] 
is_organic = []
website = [] 
desc = [] 
```

```python
#extract the values for the beer dataset

for beer in beers:
        
    #get the value of each key and if the key doesn't exist, set a variable to be None
    try: idlsNew = beer['id'] # the id in red comes from the keys on top.  (idlsNew is a variable)
    except: idlsNew = None
            
    try: nameN = beer['name']
    except: nameN = None
            
    try: yr_establishN = beer['established'] 
    except: yr_establishN = None
            
    try: statusN = beer['status']
    except: statusN = None
            
    try: is_bizN = beer['isInBusiness']
    except: is_bizN = None
            
    try: is_massN = beer['isMassOwned']
    except: is_massN = None
            
    try: is_organicN = beer['isOrganic']
    except: is_organicN = None
            
    try: websiteN = beer['website']
    except: websiteN = None
            
    try: descN = beer['description']
    except: descN = None
            
  
        # in is_biz.append(is_biz), is_biz (before the .append)is the list name
        #and is_biz in the bracket is the name of the variable in the trial
    
    idls.append(idlsNew)  # things in the bracket are the new value for the list...
    namels.append(nameN) # things in the bracket are from top where we have try:...
    yr_establish.append(yr_establishN)
    status.append(statusN)
    is_biz.append(is_bizN)
    is_mass.append(is_massN)
    is_organic.append(is_organicN)
    website.append(websiteN)
    desc.append(descN)
    
```

```python
#check number of items in each list
print(len(idls))
print(len(namels))
print(len(yr_establish))
print(len(status))
print(len(is_biz))
print(len(is_mass))
print(len(is_organic))
print(len(website))
print(len(desc))
```

```python
#match indices of each list and zip into one list
beerList = list(zip(idls, namels, yr_establish, 
                    status, is_biz, is_mass, is_organic, 
                    website, desc))

#names for columns in data frame
colnames = ['ID', 'Name', 'Year of Establishment', 'Status', 'Is it in business?', 'Is it mass owned?', 
            'Is it organic?', 'Website', 'Description']

```

```python
#create data frame with column names in it
df = pd.DataFrame(beerList, columns=colnames)

df
```
