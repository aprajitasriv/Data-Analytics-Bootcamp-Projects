```python
#Standard Deviation Method
#Delete price outside ~2 std dev (+/-1.96%)

meangrade = df_w_bath['PRICE'].mean()
stdgrade = df_w_bath['PRICE'].std()
toprange = meangrade + stdgrade * 1.96
botrange = meangrade - stdgrade * 1.96


df_w_bath = df_w_bath.drop(df_w_bath[df_w_bath['PRICE'] > toprange].index)
df_w_bath = df_w_bath.drop(df_w_bath[df_w_bath['PRICE'] < botrange].index)

df_w_bath.head()
```

```python
#drop blank data 
df=df.dropna()
```

```python
#Define bins as 0-500000, 500001-1000000, 1000001 -1500000, 1500001-2000000, 2000001-2500000
bins = [0,500000, 1000000, 1500000, 2000000, 2500000]

# Create names for the four groups
group_names = ['A', 'B', 'C', 'D', 'E']

#make new column with letter grades
df['NE_w_Price_bucket'] = pd.cut(df['PRICE'], bins,labels=group_names)
df.head()
```

```python
#boxplot for stats
sns.boxplot(data=df)
```
